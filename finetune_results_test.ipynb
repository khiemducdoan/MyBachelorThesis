```python
import matplotlib.pyplot as plt
import numpy as np

# Dữ liệu
num_heads = [1, 2, 4, 8, 16, 32]
accuracy = [80.2, 78.21, 82.2, 79.2, 78.21, 78.21]
precision = [77.86, 77.32, 79.47, 69.21, 71.48, 63.3]
f1_weighted = [78.86, 74.59, 81.83, 74.38, 74.9, 73.48]
recall = [71.77, 69.35, 78.17, 65, 67.4, 57.28]

x = np.arange(len(num_heads))
width = 0.2

plt.figure(figsize=(12, 6))
plt.bar(x - 1.5*width, accuracy, width, label='Accuracy')
plt.bar(x - 0.5*width, precision, width, label='Precision')
plt.bar(x + 0.5*width, f1_weighted, width, label='F1 Weighted')
plt.bar(x + 1.5*width, recall, width, label='Recall')

plt.xlabel('Number of Attention Heads')
plt.ylabel('Score (%)')
plt.title('Model Performance Metrics by num_heads')
plt.xticks(x, num_heads)
plt.ylim(0, 100)
plt.legend()
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Dữ liệu
num_heads = [1, 2, 4, 8, 16, 32]
metrics = {
    "Accuracy": [80.2, 78.21, 82.2, 79.2, 78.21, 78.21],
    "Precision": [77.86, 77.32, 79.47, 69.21, 71.48, 63.3],
    "F1 Weighted": [78.86, 74.59, 81.83, 74.38, 74.9, 73.48],
    "Recall": [71.77, 69.35, 78.17, 65, 67.4, 57.28]
}

labels = list(metrics.keys())
num_vars = len(labels)

# Góc cho mỗi trục
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]  # đóng vòng tròn

# Radar plot
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))

# Vẽ từng mô hình với num_heads khác nhau
for i, nh in enumerate(num_heads):
    values = [metrics[metric][i] for metric in labels]
    values += values[:1]
    ax.plot(angles, values, label=f'num_heads = {nh}')
    ax.fill(angles, values, alpha=0.1)

# Thiết lập nhãn
ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels)
ax.set_yticks([20, 40, 60, 80, 100])
ax.set_ylim(0, 100)
ax.set_title("Radar Chart of Model Performance by num_heads", y=1.08)
ax.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))

plt.tight_layout()
plt.show()
```
